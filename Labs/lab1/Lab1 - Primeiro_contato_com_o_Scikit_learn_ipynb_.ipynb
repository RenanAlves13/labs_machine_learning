# Classificação com Iris Dataset (NB)
Dataset contendo 3 tipos diferentes de íris (Setosa, Versicolour e Virginica) e 4 features: comprimento e largura de pétalas e sépalas. [Ver imagem](https://www.sobiologia.com.br/conteudos/figuras/Morfofisiologia_vegetal/flor2.jpg).

### Changing to Pandas DataFrame to better visualization

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add target column
df['label'] = iris.target

# Add column with classes names
df['species'] = df['label'].apply(lambda x: iris.target_names[x])

print(df.head())  # Show the first five lines
```

### Fazendo algumas predições

```python
x = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    x, y, random_state=42, stratify=y
)

model.fit(X_train, y_train)
y_model = model.predict(X_test)
```

```python
resultados = pd.DataFrame({'Real': y_test, 'Previsto': y_model})
print(resultados.head(15))
```
