# Classificação com KNN - Iris Dataset
Dataset contendo 3 tipos diferentes de íris (Setosa, Versicolour e Virginica) e 4 features: comprimento e largura de pétalas e sépalas. [Ver imagem](https://www.sobiologia.com.br/conteudos/figuras/Morfofisiologia_vegetal/flor2.jpg).

### Alguns Imports

```
from sklearn import datasets
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

### Carregando o dataset

```
iris = datasets.load_iris()
iris.data[:5]
```

### Preparando o classificador com um determinado valor de k

#### Imports

```
from sklearn.neighbors import KNeighborsClassifier
```

#### Instanciando
KNeighborsClassifier(...)

```
knn = KNeighborsClassifier(n_neighbors=5)
```

#### Definindo o conjunto de treino
... data[:140]

... target[:140]

```
X_train = iris.data[:140]
y_train = iris.target[:140]
```

#### Treinando com fit(...)

```
knn.fit(X_train, y_train)
```

### Avaliando a acurácia

#### Realizando a predicao para um conjunto de dados

#### Imports

```
from sklearn.metrics import accuracy_score
```

#### Calculando a acurácia
accuracy_score(rótulos de teste, rótulos preditos)

```
X_test = iris.data[140:]
y_test = iris.target[140:]

y_pred = knn.predict(X_test)

acuracia = accuracy_score(y_test, y_pred)
print(f"A acurácia do modelo nos dados de teste é: {acuracia}")
```

### Preparando novos conjuntos de treino e teste

#### Imports

```
from sklearn.model_selection import train_test_split
```

#### Definindo os conjuntos
train_test_split(valores das features, valores dos rótulos, test_size, random_state=42)

#### Treinando novamente

```
iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
knn_split = KNeighborsClassifier(n_neighbors=5)
knn_split.fit(X_train, y_train)
```

### Avaliando a acurácia

#### Imports

#### Realizando predições e calculando a acurácia
knn.predict(...) e accuracy_score(...)

```
y_pred = knn_split.predict(X_test)
acuracia = accuracy_score(y_test, y_pred)
print(f"Rótulos Verdadeiros (y_test): {y_test}")
```
